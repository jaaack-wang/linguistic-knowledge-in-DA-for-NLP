```python
from os.path import exists
```

```python
def run(net, data_size, suffix="", ep=3):
    if exists('./checkpoints'):
        !rm -r ./checkpoints
    
    train_set_suffix = f"_{data_size}{suffix}"
    
    train_set_path = './DATA/train' + train_set_suffix + '.txt' 

    !python -m paddle.distributed.launch --gpus "0" train.py --vocab_path='./vocab.txt' \
    --train_set_path={train_set_path} \
    --dev_set_path='./dev.txt' \
    --device=gpu \
    --network={net} \
    --lr=5e-4 \
    --batch_size=64 \
    --epochs={ep} \
    --save_dir='./checkpoints'

    !CUDA_VISIBLE_DEVICES=0 python predict.py --vocab_path='./vocab.txt' --train_set_suffix={train_set_suffix} --out_dir='./ablation_tests_pred/' --device=gpu --network={net} --params_path='./checkpoints/final.pdparams'
```

```python
from IPython.display import clear_output

net = 'gru'
data_size = ['5k', '10k', '25k', '50k', '75k', '100k', '125k', '150k', '175k', '200k', 'full']
suffix = ['', '_sr_reda', '_rs_reda', '_ri_reda', '_rd_reda', '_rm_reda', 
'_sr_reda_ngram', '_rs_reda_ngram', '_ri_reda_ngram', '_rd_reda_ngram', '_rm_reda_ngram']

for sfx in suffix:
    for size in data_size:
        if size in ['5k', '10k', '25k']:
            ep = 3
        elif size in ['50k', '75k', '100k']:
            ep = 2
        else:
            ep = 1

        run(net, size, sfx, ep)
        clear_output(wait=True)
```
