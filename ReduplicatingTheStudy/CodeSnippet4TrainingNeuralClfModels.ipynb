```python
from os.path import exists
```

```python
def run(net, data_size, aug_type):
    if exists('./checkpoints'):
        !rm -r ./checkpoints
    
    if aug_type == 'reda':
        train_set_suffix = f"_{data_size}_aug_reda"
    else:
        train_set_suffix = f"_{data_size}_aug_reda_ngram"
    
    train_set_path = './train' + train_set_suffix + '.txt' 

    !python -m paddle.distributed.launch --gpus "0" train.py --vocab_path='./vocab.txt' \
    --train_set_path={train_set_path} \
    --dev_set_path='./dev.txt' \
    --device=gpu \
    --network={net} \
    --lr=5e-4 \
    --batch_size=64 \
    --epochs=3 \
    --save_dir='./checkpoints'

    !CUDA_VISIBLE_DEVICES=0 python predict.py --vocab_path='./vocab.txt' --train_set_suffix={train_set_suffix} --out_dir='./tests_pred/' --device=gpu --network={net} --params_path='./checkpoints/final.pdparams'
```

```python
for n in ['bow', 'cnn', 'lstm', 'gru']:
    run(n, '50k', 'reda')
    run(n, '50k', 'reda_ngram')
```
